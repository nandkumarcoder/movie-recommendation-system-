```python
import numpy as np
import pandas as pd

movie=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
movie.head()
```

```python
movie=movie.merge(credits,on='title').
```

```python
movie = movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "SciencScience Fiction"}]
# data processing 
import ast
def convert(obj):
    li=[]
    for i in ast.literal_eval(obj):
        li.append(i['name'])
    return li
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convert2(obj):
    li=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           li.append(i['name'])
            # counter = counter+1
        else:
            break
    return li
movie['cast'].apply(convert2)
```

```python
movie.head()
```

```python
def directer_ft(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movie['crew']=movie['crew'].apply(directer_ft)
```

```python
movie.head()
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie['genres']=movie['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['cast']=movie['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['crew']=movie['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movie.head()
```

```python
movie['tags']=movie['overview']+movie['keywords']+movie['cast']+movie['crew']+movie['genres']
```

```python
movie.head()
```

```python
new_df=movie[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectores=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectores
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i  in  text.split():
        y.append(ps.stem(i))
    return " ".join(y)
           
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectores)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
def recomment(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    # distances=similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
